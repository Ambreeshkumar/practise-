```python
# create a dictionary with five fields each
datafm = {
    'A':['A1', 'nan', 'A3', 'nan', 'A5'], 
    'B':['B1', 'B2', 'B3', 'B4', 'B5'], 
    'C':['C1', 'C2', 'C3', 'C4', 'C5'], 
    'D':['D1', 'nan', 'D3', 'D4', 'D5'], 
    'E':['E1', 'E2', 'E3', 'E4', 'E5']}
```

```python
df = pd.DataFrame(datafm)
  
df.head()
```

```python
df.dropna(how = 'all', axis = 1)
```

```python
# remove duplicates;

df.drop_duplicates(subset=['A'])
```

```python
df.drop_duplicates(subset=['B'],keep=False)
```
